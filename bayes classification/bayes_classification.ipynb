```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_20newsgroups

sns.set(palette='Set2')

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
remove = ("headers", "footers")
random_state = 42

data_train = fetch_20newsgroups(
    subset="train", shuffle=True, random_state=random_state, remove=remove
)
data_test = fetch_20newsgroups(
    subset="test", shuffle=True, random_state=random_state, remove=remove
)
```

```python
train_size = int(len(data_train.data) * 0.7)
val_size = int(len(data_train.data) * 0.15)

train_data = data_train.data[:train_size]
train_target = data_train.target[:train_size]

val_data = data_train.data[train_size:train_size+val_size]
val_target = data_train.target[train_size:train_size+val_size]

test_data = data_test.data
test_target = data_test.target
```

```python
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_data)
X_val = vectorizer.transform(val_data)

nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, train_target)
```

```python
val_predictions = nb_classifier.predict(X_val)
val_accuracy = accuracy_score(val_target, val_predictions)

print("Точность на валидации:", val_accuracy)
```

```python
X_test = vectorizer.transform(test_data)

test_predictions = nb_classifier.predict(X_test)
test_accuracy = accuracy_score(test_target, test_predictions)

print("Точность на тестовой выборке:", test_accuracy)
```
